```python
plt.plot(x,y)
```

```python
sns.lineplot(x=x,y=y)
```

### Pie chart

```python
data = [2,6,5,10]
plt.pie(data)
plt.show()
```

```python
data = [2,6,5,10]
plt.pie(data,[0,0,0.06,0.08],labels=['t1','t2','t3','t4'],autopct = "%1.1f%%",shadow = True, colors=['red','yellow','pink','purple'])
plt.title('Pie Diagram')
plt.show()
```
